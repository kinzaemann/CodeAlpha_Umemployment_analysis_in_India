```python
import string 
para1=""" python !!is the most popular, language in AI and, data-science; it supports Libraries 
like Numpy , pandas , and Tensorflow ! Many students learn it at AIR university ."""
def cleantext(text):
    text=text.strip()
    text=text.lower()
    text = text.translate(str.maketrans('', '', string.punctuation))
    text = " ".join(text.split())
    text = text.capitalize()
    text[0].upper() + text[1:]
    text = text.rstrip(".") + "."
    return text

clean_para1=cleantext(para1)
print(clean_para1)
```
